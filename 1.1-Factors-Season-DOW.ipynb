```python
import numpy as np
import os
import pandas as pd

dirWork = os.getcwd()
dirData = os.path.join(dirWork, r'data')
dirCCS  = os.path.join(dirData, r'udot-ccs')
dirCCS  = os.path.join(dirData, r'udot-ccs')
dirResults = os.path.join(dirWork, r'data-processed\factors')

print("Working Directory: " + dirWork)
print("Data Directory: " + dirData)
print("CCS Directory: " + dirCCS)

#which site grouping to use
facgeogroup_use = "COFACGEO"
```

# Season Groups

```python
df_StationGroups = pd.read_csv(os.path.join(dirCCS, r'station_group_to_facgroup.csv'))
df_StationGroups = df_StationGroups[df_StationGroups['FACGEOGROUP']==facgeogroup_use]

#default
df_StationGroups['FTCOMBOGROUP'] = 'ArterialExpressway'
df_StationGroups['FACGEO'] = ''

#split up FACGRP value into two fields
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('ArterialExpressway'),'FTCOMBOGROUP'] = 'ArterialExpressway'
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('ArterialExpressway'),'FACGEO'] = df_StationGroups['FACGRP'].str.replace('ArterialExpressway', '', regex=False)
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('Freeway'),'FTCOMBOGROUP'] = 'Freeway'
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('Freeway'),'FACGEO'] = df_StationGroups['FACGRP'].str.replace('Freeway', '', regex=False)
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('Arterial') & -(df_StationGroups['FACGRP'].str.contains('Expressway')),'FTCOMBOGROUP'] = 'Arterial'
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('Arterial') & -(df_StationGroups['FACGRP'].str.contains('Expressway')),'FACGEO'] = df_StationGroups['FACGRP'].str.replace('Arterial', '', regex=False)
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('Expressway') & -(df_StationGroups['FACGRP'].str.contains('Arterial')),'FTCOMBOGROUP'] = 'Expressway'
df_StationGroups.loc[df_StationGroups['FACGRP'].str.contains('Expressway') & -(df_StationGroups['FACGRP'].str.contains('Arterial')),'FACGEO'] = df_StationGroups['FACGRP'].str.replace('Expressway', '', regex=False)

df_StationGroups.loc[df_StationGroups['FACGEO']=='','FACGEO'] = df_StationGroups['FACGRP']

#created statewide freeway group XX3 messing up others, so drop it
df_StationGroups = df_StationGroups[df_StationGroups['STATIONGROUP']!='XX3']

df_StationGroups
```

## Area Types

```python
df_AT = pd.read_csv(os.path.join(dirCCS, r'area_type_group_to_area_type.csv'))
df_AT
```

```python
df_StationGroups_AT = pd.DataFrame.merge(df_StationGroups,df_AT,on='FACATGROUP',how='left')
df_StationGroups_AT
```

## Volume Groups

```python
df_StationGroups_AT.loc[(df_StationGroups_AT['FACVOLCLASS'] == '0-AllVolumes'), 'VOLUMEGROUP'] = "All"

#rename Vol Class groups for easy join
df_StationGroups_AT.loc[(df_StationGroups_AT['FACGEO'] == 'Southeast') & (df_StationGroups_AT['FACVOLCLASS'] == '1-LowVolume' ), 'VOLUMEGROUP'] = "SERuralLow"
df_StationGroups_AT.loc[(df_StationGroups_AT['FACGEO'] == 'Southeast') & (df_StationGroups_AT['FACVOLCLASS'] == '2-MidVolume' ), 'VOLUMEGROUP'] = "SERuralMid"

df_StationGroups_AT.loc[(df_StationGroups_AT['FACGEO'] == 'Statewide') & (df_StationGroups_AT['FACVOLCLASS'] == '1-LowVolume' ), 'VOLUMEGROUP'] = "UTRuralLow"
df_StationGroups_AT.loc[(df_StationGroups_AT['FACGEO'] == 'Statewide') & (df_StationGroups_AT['FACVOLCLASS'] == '2-MidVolume' ), 'VOLUMEGROUP'] = "UTRuralMid"
df_StationGroups_AT.loc[(df_StationGroups_AT['FACGEO'] == 'Statewide') & (df_StationGroups_AT['FACVOLCLASS'] == '3-HighVolume'), 'VOLUMEGROUP'] = "UTRuralHigh"

df_StationGroups_AT
```

```python
df_Volume = pd.read_csv(os.path.join(dirCCS, r'volume_groups.csv'))
df_Volume
```

```python
df_StationGroups_ATVol = pd.DataFrame.merge(df_StationGroups_AT,df_Volume,on='VOLUMEGROUP')
pd.set_option('display.max_rows', df_StationGroups_ATVol.shape[0]+1)
df_StationGroups_ATVol
```

```python
df_Groups = df_StationGroups_ATVol

df_Groups['SG_FG_Arterial'] = df_Groups['STATIONGROUP']
df_Groups['SG_FG_Freeway']  = df_Groups['STATIONGROUP']
df_Groups['SG_FG_Managed']  = df_Groups['STATIONGROUP']
df_Groups['SG_FG_CD_Road']  = df_Groups['STATIONGROUP']

#df_Groups = df_Groups.drop(columns=[,'FACATGROUP','FACVOLCLASS'])

df_Groups
```

```python
df_Groups[df_Groups['FACGEO']=='Statewide']
```

# Functional Groups

```python
#custom changes
#see SiteGroupSeasonFactors.xlsx for SiteGroups and Factors

#Provide for conversion of arterial segment to freeway. provide with XX4, which is generic commuter freeway (average of I-215, Legacy, and SR-201)
#assume new freeways are always like a commuter freeway and not long-haul
df_Groups.loc[(df_Groups['FACGEO'] == 'Statewide'), 'SG_FG_Freeway'] = "XX3"
df_Groups.loc[(df_Groups['FACGEO'] == 'Statewide'), 'SG_FG_Managed'] = "XX3"
df_Groups.loc[(df_Groups['FACGEO'] == 'Statewide'), 'SG_FG_CD_Road'] = "XX3"

df_Groups.loc[(df_Groups['FACGEO'] == 'Southeast'), 'SG_FG_Freeway'] = "XX3"
df_Groups.loc[(df_Groups['FACGEO'] == 'Southeast'), 'SG_FG_Managed'] = "XX3"
df_Groups.loc[(df_Groups['FACGEO'] == 'Southeast'), 'SG_FG_CD_Road'] = "XX3"

#set arterials on freeway segments to statewide arterial SiteGroup
df_Groups.loc[(df_Groups['FTCOMBOGROUP'].str.contains('Freeway')) & (df_Groups['AREATYPE'] == 1), 'SG_FG_Arterial'] = "COQ"
df_Groups.loc[(df_Groups['FTCOMBOGROUP'].str.contains('Freeway')) & (df_Groups['AREATYPE'] == 2), 'SG_FG_Arterial'] = "COS"
df_Groups.loc[(df_Groups['FTCOMBOGROUP'].str.contains('Freeway')) & (df_Groups['AREATYPE'] == 3), 'SG_FG_Arterial'] = "COT"
df_Groups.loc[(df_Groups['FTCOMBOGROUP'].str.contains('Freeway')) & (df_Groups['AREATYPE'] == 4), 'SG_FG_Arterial'] = "COU"
df_Groups.loc[(df_Groups['FTCOMBOGROUP'].str.contains('Freeway')) & (df_Groups['AREATYPE'] == 5), 'SG_FG_Arterial'] = "COU"


#could do more with other SsnGrp

df_Groups
```

```python
df_Groups[df_Groups['FACGEO']=='Statewide']
```

```python
df_Groups_melt = pd.melt(df_Groups, id_vars=['FACGEO','AREATYPE','VOLUMEFROM','VOLUMETO'], value_vars=['SG_FG_Arterial','SG_FG_Freeway','SG_FG_Managed','SG_FG_CD_Road'])
df_Groups_melt.columns = ('FACGEO','AREATYPE','VOLUMEFROM','VOLUMETO','FUNCGROUP','STATIONGROUP')
df_Groups_melt['FUNCGROUP'] = df_Groups_melt['FUNCGROUP'].str.replace("SG_FG_", "")
df_Groups_melt['FUNCGROUP'] = df_Groups_melt['FUNCGROUP'].str.replace("_", " ")
df_Groups_melt
```

```python
df_Groups_melt[df_Groups_melt['FACGEO']=='Statewide']
```

```python
df_Groups_melt.to_csv(os.path.join(dirResults,r'facgeo_fungroup_areatype_volume_to_station_group.csv'),index=False)
```

# Site Group Season/DOW Factors

```python
#read excel with season factors
df_Factors = pd.read_csv(os.path.join(dirData, r'CCSFactors.csv'))
df_Factors['FAC_ANN'] = 1.00
df_Factors
```

```python
df_Factors_melt = pd.melt(df_Factors, id_vars=['STATIONGROUP'], value_vars=['FAC_JAN','FAC_FEB','FAC_MAR','FAC_APR','FAC_MAY','FAC_JUN','FAC_JUL','FAC_AUG','FAC_SEP','FAC_OCT','FAC_NOV','FAC_DEC','FAC_WIN','FAC_SPR','FAC_SUM','FAC_FAL','FAC_MAXMO','FAC_ANN','FAC_MON','FAC_TUE','FAC_WED','FAC_THU','FAC_FRI','FAC_SAT','FAC_SUN','FAC_WDAVG','FAC_WEAVG'])
#['M01-Jan','M02-Feb','M03-Mar','M04-Apr','M05-May','M06-Jun','M07-Jul','M08-Aug','M09-Sep','M10-Oct','M11-Nov','M12-Dec','S01-Win','S02-Spr','S03-Sum','S04-Fal','MaxMonth','MaxSeason','S00-Ann'])
df_Factors_melt.columns = ['STATIONGROUP','FACNAME','FACTOR']
df_Factors_melt
```

```python
df_Factors_melt
```

```python
df_season_group_field_names = pd.read_csv(os.path.join(dirCCS, r'season_group_field_names.csv'))
df_dow_group_field_names    = pd.read_csv(os.path.join(dirCCS, r'dow_group_field_names.csv'   ))

df_SeasonFactors = pd.DataFrame.merge(df_Factors_melt, df_season_group_field_names,left_on='FACNAME',right_on='SGFLDNM')
df_SeasonFactors = df_SeasonFactors.drop(columns=['FACNAME','SGFLDNM'])

display(df_SeasonFactors)

df_DOWFactors = pd.DataFrame.merge(df_Factors_melt, df_dow_group_field_names,left_on='FACNAME',right_on='DOWFLDNM')
df_DOWFactors = df_DOWFactors.drop(columns=['FACNAME','DOWFLDNM','DAYFLAG','DAYNAME'])

display(df_DOWFactors)
```

# Export CSV

```python
df_SeasonFactors.to_csv(os.path.join(dirResults,r'SeasonFactors.csv'),index=False)
df_DOWFactors.to_csv(os.path.join(dirResults,r'DOWFactors.csv'),index=False)
display('CSVs Exported')
```
